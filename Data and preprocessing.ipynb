```python

# Define the directory where your files are located
directory_path_house = "Data/House Hearings"

house_documents = []

# Loop through each file in the directory
for filename in os.listdir(directory_path_house):
    if filename.endswith('.txt'):  # Check if the file is a text file
        file_path = os.path.join(directory_path_house, filename)
        
        # Read the file
        with open(file_path, 'r', encoding='utf-8') as file:
            file_content = file.read()

        # Parse the HTML content
        soup = BeautifulSoup(file_content, 'html.parser')

        # Extract the text
        house_text = soup.get_text()

        # Append the text to the documents list
        # The ID will be assigned later, based on the DataFrame's index
        house_documents.append({
            'text': house_text,
            'congress': 'house'
        })

# Create a DataFrame
df_house_documents = pd.DataFrame(house_documents)

# Assign the row number as the 'id'
df_house_documents['id'] = df_house_documents.index

# Add a new column 'Period' to the DataFrame and set its value for all rows to '2023-2024'
df_house_documents['period'] = '2023-2024'

# Show the DataFrame to verify its structure
df_house_documents
```

### Preprocesing

#### Eliminate the information above the first opening statement (e.g. information about the commisioners)

```python
# Define a function to remove text before the first "OPENING STATEMENT" more efficiently
def efficient_remove_text(text):
    opening_statement_index = text.find("OPENING STATEMENT")
    if opening_statement_index != -1:
        return text[opening_statement_index:]
    else:
        return text

# Apply the new function to each row in the 'text' column
all_documents['text'] = all_documents['text'].apply(efficient_remove_text)


all_documents
```

```python
# Define a function to remove text before "The committee met," or "The subcommittee met,"
def remove_text_before_meeting_statements(text):
    patterns = ["The committee met,", "The subcommittee met,"]
    earliest_index = len(text)
    
    for pattern in patterns:
        index = text.find(pattern)
        if index != -1 and index < earliest_index:
            earliest_index = index

    if earliest_index < len(text):
        return text[earliest_index:]
    else:
        return text

# Apply the function to each row in the 'text' column
all_documents['text'] = all_documents['text'].apply(remove_text_before_meeting_statements)

all_documents
```

#### Tokenization, removing stop words and lemmatization

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text_v2(text):
    """
    Function to tokenize, convert to lowercase, remove punctuation and stopwords, 
    and perform lemmatization on a given text.
    """
    # Lowercasing
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenization
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    return lemmatized_tokens


# Applying the updated preprocessing to the 'text' column
all_documents['processed_text_v2'] = all_documents['text'].apply(preprocess_text_v2)

# Optionally, save the processed data to a new CSV file
all_documents.to_csv('processed_data.csv', index=False)
all_documents
```
